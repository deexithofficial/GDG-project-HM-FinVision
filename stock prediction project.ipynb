```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://www.screener.in/company/NIFTY/"
```

```python
headers = {
    
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36"

}
```

```python
response = requests.get(url, headers=headers)
```

```python
response
```

```python
response.raise_for_status()
```

```python
soup = BeautifulSoup(response.text, "html.parser")
```

```python
table = soup.find("table" , class_="data-table text-nowrap striped mark-visited").text.strip()
```

```python
table_body = soup.select_one("#constituents > div.responsive-holder.fill-card-width > table > tbody")
```

```python
if table_body:
    # Extract rows and columns
    rows = table_body.find_all("tr")
    data = []
    for row in rows:
        row_data = [cell.get_text(strip=True) for cell in row.find_all("td")]
        data.append(row_data)
        
    # Define column headers (update based on actual table structure)
    columns = ["S.NO", "Name", "CMP", "Mar Cap", "Div Yield", "Net Profit", 
                "Qtr profit", "LTP", "Sales Qtr", "Volume", "% Change"]
        
    # Convert to DataFrame
    df = pd.DataFrame(data, columns=columns)
    print(df)
else:
    print("Table body not found. Check the selector or website structure.")
```
