```python
def black_scholes(S, K, T, r, sigma, option_type='call'):
    d1 = (np.log(S / K) + (r + 0.5 * sigma ** 2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    
    if option_type == 'call':
        return S * norm.cdf(d1) - K * np.exp(-r * T) * norm.cdf(d2)
    elif option_type == 'put':
        return K * np.exp(-r * T) * norm.cdf(-d2) - S * norm.cdf(-d1)
```

```python
S_values = merged_df['Close']
K = merged_df['strike']
```

```python
K
```

```python
S_values
```

```python
merged_df['theoretical_call_prices'] = black_scholes(
    S=merged_df['Close'],
    K=merged_df['strike'],
    T=30/365,
    r=0.05,  # Assuming openInterest is used as the risk-free rate
    sigma=merged_df['Returns'].std() * np.sqrt(252),
    option_type='call'
)
```

```python
merged_df
```

```python
plt.figure(figsize=(10, 6))

plt.plot(merged_df['strike'], merged_df['theoretical_call_prices'], label='Theoretical Premium', marker='o', linestyle='--')
plt.plot(merged_df['strike'], merged_df['lastPrice'], label='Calculated Premium', marker='x', linestyle='-')
plt.title('Option Premiums vs. Strike Prices for the expiration date of 2024-12-30', fontsize=14)
plt.xlabel('Strike Price', fontsize=12)
plt.ylabel('Premium', fontsize=12)
plt.legend()
plt.grid(True)

plt.show()
```
